```python
#Its time to prep for getting into my customer segmentation. I plan on using the k-means clustering method. To find best way to do this i will
#be using the elbow method and the silhouette method to see how many clusters should be made.
features = ['MntTotal', 'Income', 'Age', 'education_2n Cycle', 'education_Basic', 'education_Graduation', 
            'education_Master', 'education_PhD', 'marital_Divorced', 'marital_Married', 'marital_Single', 
            'marital_Together', 'marital_Widow', 'Kidhome', 'Teenhome']

# Extract the selected features from the DataFrame
data = df[features].copy()  # Create a copy of the subset DataFrame

# Drop rows with missing values
data.dropna(inplace=True)

# Standardize the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Define the range of clusters to consider
min_clusters = 2
max_clusters = 10

# Initialize lists to store results
inertia_values = []
silhouette_scores = []

# Iterate over different numbers of clusters
for k in range(min_clusters, max_clusters + 1):
    # Apply K-means clustering
    kmeans = KMeans(n_clusters=k, n_init='auto', random_state=42)
    kmeans.fit(scaled_data)
    
    # Compute inertia (within-cluster sum of squares)
    inertia_values.append(kmeans.inertia_)
    
    # Compute silhouette score
    silhouette_scores.append(silhouette_score(scaled_data, kmeans.labels_))

# Plot Elbow Method
plt.figure(figsize=(10, 5))
plt.plot(range(min_clusters, max_clusters + 1), inertia_values, marker='o', linestyle='-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xticks(range(min_clusters, max_clusters + 1))
plt.grid(True)
plt.show()

# Plot Silhouette Score
plt.figure(figsize=(10, 5))
plt.plot(range(min_clusters, max_clusters + 1), silhouette_scores, marker='o', linestyle='-')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal Number of Clusters')
plt.xticks(range(min_clusters, max_clusters + 1))
plt.grid(True)
plt.show()
```

```python
#So both methods gave me two different answers so I will be comparing the answers of both to see what fits better.

# Extract the selected features from the DataFrame
data = df[features].copy()  # Create a copy of the subset DataFrame

# Drop rows with missing values
data.dropna(inplace=True)

# Define the number of clusters to compare
cluster_numbers = [5, 6]

# Initialize subplots
fig, axes = plt.subplots(nrows=1, ncols=len(cluster_numbers), figsize=(12, 6))
if len(cluster_numbers) == 1:
    axes = [axes]  # Ensure axes is iterable

# Iterate over cluster numbers
for i, k in enumerate(cluster_numbers):
    # Apply K-means clustering
    kmeans = KMeans(n_clusters=k, random_state=42)
    clusters = kmeans.fit_predict(data)
    
    # Plot scatter plot with cluster colors
    ax = axes[i]
    ax.scatter(data.iloc[:, 0], data.iloc[:, 1], c=clusters, cmap='viridis', marker='o', alpha=0.5)
    ax.set_title(f'{k} Clusters')
    ax.set_xlabel(features[0])
    ax.set_ylabel(features[1])

plt.tight_layout()
plt.show()
```

```python
# Extract the selected features from the DataFrame
X = df[features].copy()
# Drop rows with missing values
X.dropna(inplace=True)
num_clusters = 5 
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(X)
df['Cluster'] = cluster_labels
cluster_centroids = pd.DataFrame(kmeans.cluster_centers_, columns=features)
print(cluster_centroids)
plt.show()
```
