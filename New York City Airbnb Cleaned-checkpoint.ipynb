```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_excel(r"C:\Users\cc225\OneDrive\Documents\AB_NYC_2019.xlsx")
df
```

```python
df= df.drop_duplicates()
df
```

```python
df.info()
```

```python
df['neighbourhood_group']= df['neighbourhood_group'].astype('category')
df['neighbourhood']= df['neighbourhood'].astype('category')
df['room_type']= df['room_type'].astype('category')
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna({'name':'N/A'}, inplace=True)
df.drop(['host_name', 'last_review'], axis=1, inplace=True)
df.fillna({'reviews_per_month': 0}, inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

```python
df_numeric = df.select_dtypes(include=[np.number])
for col in df_numeric.columns:
    z_score = np.abs(stats.zscore(df[col]))
    outliers_num = len(np.where(z_score > 3)[0])
    if outliers_num:
        print('{}: {}'.format(col, outliers_num))
```

```python
z_scores = np.abs(stats.zscore(df_numeric))
df= df[(z_scores < 3).all(axis=1)]
df.shape
df.describe()
```
