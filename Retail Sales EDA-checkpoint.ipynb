```python
#I can see that the date, product category, and gender are saved under the object which would be the incorrect dtype.
df['Date'] = pd.to_datetime(df['Date'])
df['Product Category'] = df['Product Category'].astype('category')
df['Gender'] = df['Gender'].astype('category')
df.info()
```

```python
#Transaction ID and Customer ID do not seem necessary to keep for my analysis so i will be removing it.
df.drop(['Transaction ID'], axis=1, inplace=True)
df.drop(['Customer ID'], axis=1, inplace=True)
df.head()
```

```python
outlier_cols = ['Price per Unit', 'Total Amount']
```

```python
#I am plotting the data to search for outliers
for col in outlier_cols:
    plt.figure(figsize=(5, 5))
    sns.boxplot(x=df[col])
    plt.show()
```

```python
for col in outlier_cols:
    z_score = np.abs(stats.zscore(df[col]))
    outliers_num = len(np.where(z_score > 3)[0])
    if outliers_num:
        print('{}: {}'.format(col, outliers_num))
    else:
        print('{}: No outliers detected'.format(col))
```

```python
df.set_index('Date', inplace=True)
print(df.index)
```

```python
# Set the index to datetime
df.index = pd.to_datetime(df.index)

# Pivot the DataFrame to get monthly totals for each product category
pivot_table = df.pivot_table(index=df.index.month, columns='Product Category', values='Total Amount', aggfunc='sum')

# Plotting
colors = ['skyblue', 'salmon', 'lightgreen']  # Define colors for each category
pivot_table.plot(kind='bar', figsize=(10, 8), color=colors)
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.title('Monthly Sells Data')
plt.legend(title='Product Category')
plt.show()
```

```python
# Plotting
colors = ['skyblue', 'salmon', 'lightgreen']  # Define colors for each category
pivot_table.plot(kind='line', figsize=(12, 6), color=colors)
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.title('Monthly Sells Data')
plt.legend(title='Product Category')
plt.show()
```
